```python
with open('corpus.sentence.txt', 'r', encoding='utf-8') as sentence:  # 打开语料，获得需要分词的文本
    lst = []
    while True:
        line = sentence.readline()
        if line == '':
            break
        else:
            lst.append(line[:-1])  # 将语料存入列表

with open('corpus.dict.txt', 'r', encoding='utf-8') as d:  # 打开字典文件
    dic = []
    while True:
        line = d.readline()
        if line == '':
            break
        else:
            dic.append(line[:-1])  # 将字典内容存入列表

num = int(dic[0].split('\t')[1])  # 获取最大长度

with open('corpus.out.txt', 'w', encoding='utf-8') as out:  # 创建.out文件
    out_lst = []

    with open('corpus.answer.txt', 'r', encoding='utf-8') as answer:  # 打开answer文件
        ans_lst = []
        while True:  # 读入数据
            line = answer.readline()
            if line == '':
                break
            else:
                ans_lst.append(line[:-1].split())  # 以空格隔开

    for item in lst:  # 循环语料库中每一句话
        start = 0  # 设置启示位置
        count = 10  # 设置最大长度
        ans = []
        while count >= 1:  # 设置内循环
            if count == start:
                break
            res = item[start:count]  # 截取当前长度字符串
            if res in dic:  # 判断是否存在在字典中
                ans.append(res)
                start = count
                if 10 + start < len(item):  # 防止下标访问越界
                    count = 10 + start  # 重设长度
                else:
                    count = len(item)
            else:
                if count == start + 1:  # 单个字符时处理
                    ans.append(item[start:count])
                    start = count
                    if 10 + start < len(item):
                        count = 10 + start
                    else:
                        count = len(item)
                else:  # 长度减一
                    count -= 1

        out_lst.append(ans)  # 将分词结果全部存入列表以备统计总数
        result = ' '.join(ans)
        out.write(result + '\n')  # 将分词结果写入.out文件并换行
        
with open('corpus.PRF.txt', 'w', encoding='utf-8') as PRF:
    # 统计每行的out_num, right_num, ans_num
    for i in range(0, len(out_lst)):
        
        out_num = 0
        right_num = 0
        for j in out_lst[i]:
            out_num += 1
            if j in ans_lst[i]:
                right_num += 1
                
        ans_num = 0
        for k in ans_lst[i]:
            ans_num += 1
            
        precision = right_num / out_num  # 计算正确率
        recall = right_num / ans_num  # 计算召回率
        f = precision * recall * 2 / (precision + recall)  # 计算F值
        
        PRF.write('Precision = ' + str(right_num) + '/' + str(out_num) + '=' + '{0:.2f}'.format(
            precision * 100) + '%' + '   ')
        PRF.write('Recall ='  + str(right_num) + '/' + str(out_num) + ' = ' + '{0:.2f}'.format(
            recall * 100) + '%' + '   ')
        PRF.write('F值 = ' + '{0:.2f}'.format(f * 100) + '%')
        PRF.write('\n')
```
